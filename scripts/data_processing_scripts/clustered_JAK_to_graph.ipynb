```python
import os
import pickle as pk
import matplotlib.pyplot as plt
```

```python
root = os.path.join("..", "..", "data", "JAK")
file_ = os.path.join(root, "filtered_JAK1_clusters.pk")
```

```python
with open(file_, "rb") as f:
    clusters = pk.load(f)
```

```python
sum = 0
for i, c in clusters.items():
    sum += len(c)
    print(i, len(c))
print(sum)
```

```python
cluster_sizes = sorted(list(map(len, clusters.values())), reverse=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(range(len(cluster_sizes)), cluster_sizes)
ax.set(xlabel="Clusters", ylabel="Counts", title="JAK1 Cluster Distribution")
fig.savefig(os.path.join(root, "JAK1_cluster_distribution.png"), dpi=300)
fig.show()
```
