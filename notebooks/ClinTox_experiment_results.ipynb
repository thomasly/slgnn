```python
import os
import pickle as pk

import matplotlib.pyplot as plt
```

```python
root = os.path.join("..", "logs", "GIN", "20200602_070944")

group = ["22", "23"]
folder = "ZINC10k"

with open(os.path.join(root, f"{folder}", group[0], "classifier_metrics.pk"), "rb") as f:
    met1 = pk.load(f)

with open(os.path.join(root, f"{folder}", group[1], "classifier_metrics.pk"), "rb") as f:
    met2 = pk.load(f)
```

```python
datasets = "ZINC10k"
label = "1to9"
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(range(len(met1["validating_roc_auc"])), met1["validating_roc_auc"], label="val_roc_auc_wo_encoder")
ax.plot(range(len(met2["validating_roc_auc"])), met2["validating_roc_auc"], label="val_roc_auc_encoder")
ax.set(ylim=[0, 1], xlabel="epochs", ylabel="ROC-AUC")
ax.legend()
fig.savefig(f"{datasets}_ClinTox_{label}_valrocauc.png", dpi=300)
# fig.savefig("ZINC11kJAK2_JAK2_9to1_valacc.png", dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(range(len(met1["validating_f1"])), met1["validating_f1"], label="val_f1_wo_encoder")
ax.plot(range(len(met2["validating_f1"])), met2["validating_f1"], label="val_f1_encoder")
ax.set(ylim=[0, 1], xlabel="epochs", ylabel="F1")
ax.legend()
fig.savefig(f"{datasets}_ClinTox_{label}_valf1.png", dpi=300)
# fig.savefig("ZINC11kJAK2_JAK2_9to1_valacc.png", dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(range(len(met1["validating_accuracy"])), met1["validating_accuracy"], label="val_acc_wo_encoder")
ax.plot(range(len(met2["validating_accuracy"])), met2["validating_accuracy"], label="val_acc_encoder")
ax.set(ylim=[0, 1], xlabel="epochs", ylabel="Accuracy")
ax.legend()
fig.savefig(f"{datasets}_ClinTox_{label}_valacc.png", dpi=300)
# fig.savefig("ZINC11kJAK2_JAK2_9to1_valacc.png", dpi=300)
plt.show()
```
