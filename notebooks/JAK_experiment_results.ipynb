```python
import os
import pickle as pk

import matplotlib.pyplot as plt
```

```python
root = os.path.join("..", "logs", "GIN", "20200526_012953")
encoder_data = ["ZINC1k", "ZINC1k_JAK3", "ZINC10k", "ZINC10k_JAK3"]
```

```python
group = ["0", "1"]
```

```python
with open(os.path.join(root, "ZINC1k_JAK3", group[0], "classifier_losses_accs.pk"), "rb") as f:
    met1 = pk.load(f)

with open(os.path.join(root, "ZINC1k_JAK3", group[1], "classifier_losses_accs.pk"), "rb") as f:
    met2 = pk.load(f)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(range(len(met1[0]["validating_accs"])), met1[0]["validating_accs"], label="val_acc_wo_encoder")
ax.plot(range(len(met2[0]["validating_accs"])), met2[0]["validating_accs"], label="val_acc_encoder")
ax.set(ylim=[0, 1], xlabel="epochs", ylabel="Accuracy")
ax.legend()
fig.savefig("ZINC10kJAK3_JAK3Presplitted_valacc.png", dpi=300)
# fig.savefig("ZINC11kJAK2_JAK2_9to1_valacc.png", dpi=300)
plt.show()
```
